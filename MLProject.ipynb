```python
from sklearn.preprocessing import LabelEncoder
```

```python
for col in ['gender', 'country', 'product_category', 'payment_type', 'device_type', 'referral_source','order_date']:
    le=LabelEncoder()
    data[col] = le.fit_transform(data[col])
```

```python
x=data.iloc[:,:19]
```

```python
x.head()
```

```python
y=data.iloc[:,19]
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
ytrain.value_counts()
```

```python
from sklearn.metrics import accuracy_score,precision_score,confusion_matrix,recall_score,f1_score
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf =RandomForestClassifier()
```

```python
rf.fit(xtrain,ytrain)
```

```python
ypred2=rf.predict(xtest)
ypred2
```

```python
score2=accuracy_score(ytest,ypred2)
score2
```

```python
precision2=precision_score(ytest,ypred2)
precision2
```

```python
call2=recall_score(ytest,ypred2)
call2
```

```python
f1_2=f1_score(ytest,ypred2)
f1_2
```
